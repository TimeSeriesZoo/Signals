# Time Series

Animated visualizers for time series.

```python
from neurodsp.sim import sim_oscillation, sim_powerlaw, sim_combined
from neurodsp.utils import set_random_seed
```

```python
# Import local code
from vcode.build.timeseries import build_timeseries
```

## Setup

### Settings

```python
import seaborn as sns
sns.set_context('talk')
```

```python
# Set random seed
set_random_seed(505)
```

```python
# Simulation settings
n_seconds = 100
fs = 1000
```

```python
# Notebook settings
SAVE = False
N_BUILD = 200
```

## Make Visualizers

#### Oscillating signal

```python
sig_osc = sim_oscillation(n_seconds, fs, freq=10)
```

```python
build_timeseries(sig_osc, fs, n_build=N_BUILD, save=SAVE, label='01-osc')
```

#### White Noise

```python
sig_wh = sim_powerlaw(n_seconds, fs, 0, f_range=(0.5, None))
```

```python
build_timeseries(sig_wh, fs, n_build=N_BUILD, save=SAVE, label='01-ap_wh')
```

### Pink Noise

```python
sig_pi = sim_powerlaw(n_seconds, fs, -1, f_range=(0.5, None))
```

```python
build_timeseries(sig_pi, fs, n_build=N_BUILD, save=SAVE, label='01-ap_pi')
```

### Brown Noise

```python
sig_br = sim_powerlaw(n_seconds, fs, -2, f_range=(0.5, None))
```

```python
build_timeseries(sig_br, fs, n_build=N_BUILD, save=SAVE, label='01-ap_br')
```

### Combined Signal

```python
comp_vars = [0.25, 1]
comps = {'sim_powerlaw' : {'exponent' : -1, 'f_range' : (0.5, None)},
         'sim_oscillation' : {'freq' : 10}}
sig_comb = sim_combined(n_seconds, fs, comps, component_variances=comp_vars)
```

```python
build_timeseries(sig_comb, fs, n_build=N_BUILD, save=SAVE, label='01-comb')
```

### Combined Bursty Signal

```python
comp_vars = [0.25, 1]
comps = {'sim_powerlaw' : {'exponent' : -1, 'f_range' : (0.5, None)},
         'sim_bursty_oscillation' : {'freq' : 10}}
sig_comb_burst = sim_combined(n_seconds, fs, comps, component_variances=comp_vars)
```

```python
build_timeseries(sig_comb_burst, fs, n_build=N_BUILD, save=SAVE, label='01-comb_burst')
```
